```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('penguins.csv')
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
sorted(data['flipper_length_mm'])
```

```python
data1=data['flipper_length_mm']
```

```python
quantile1,quantile3= np.percentile(data1,[25,75])
print(quantile1,quantile3)
```

```python
iqr_value=quantile3-quantile1
print(iqr_value)
```

```python
lower_bound= quantile1-(1.5*iqr_value)
upper_bound=quantile3+(1.5*iqr_value)
print(lower_bound,upper_bound)
```

```python
print(data[data1>upper_bound])
```

```python
print(data[data1<lower_bound])
```

```python
penguins= data[(data1<upper_bound)& (data1>lower_bound)]
```

```python
penguins.head()
```

```python
sorted(penguins['flipper_length_mm'])
```

```python
penguins1=pd.get_dummies(penguins,columns =['sex']).drop("sex_.",axis=1)

penguins1

```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(penguins1)
print(X)
```

```python
peng=pd.DataFrame(data=X, columns= penguins1.columns)
peng.head()
```

from sklearn.cluster import KMeans


```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(peng)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("No.of clusters")
plt.ylabel("wcss")
plt.show()
           
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(peng)
print(y_predict)
```

```python
plt.scatter(peng[y_predict==0,0],peng[y_predict==0,1],s=100,c="red",label='cluster1')
plt.scatter(peng[y_predict==1,0],peng[y_predict==1,1],s=100,c="green",label='cluster2')
plt.scatter(peng[y_predict==2,0],peng[y_predict==2,1],s=100,c="yellow",label='cluster3')
plt.scatter(peng[y_predict==3,0],peng[y_predict==3,1],s=100,c="blue",label='cluster4')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c="magenta",label="centroid")
plt.show()
```
